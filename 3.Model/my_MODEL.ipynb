```python
import os
import re
import json
import ast
import pandas as pd
import numpy as np
import torch
from transformers import PreTrainedTokenizerFast, BartForConditionalGeneration, Trainer, TrainingArguments, DataCollatorForSeq2Seq
from torch.utils.data import Dataset

# ==============================
# 1. 데이터 전처리 모듈
# ==============================
class DataProcessor:
    def __init__(self):
        pass

    def clean_text(self, text):
        if isinstance(text, str):
            text = re.sub(r'\s+', ' ', text).strip()
        else:
            text = ""
        return text

    def load_train_data(self, filepath):
        df = pd.read_csv(filepath)
        for col in ['question', 'answer', 'PDF 내 관련 문장']:
            if col in df.columns:
                df[col] = df[col].apply(self.clean_text)
        return df

# ==============================
# 2. KoBART 파인튜닝 데이터셋
# ==============================
class KoBARTDataset(Dataset):
    def __init__(self, data, tokenizer, max_length=512):
        self.tokenizer = tokenizer
        self.data = data
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        # 데이터를 가져옴
        row = self.data.iloc[idx]
        context = row['PDF 내 관련 문장'] if pd.notna(row['PDF 내 관련 문장']) else ""  # 문맥 정보 가져오기

        # input_text에 문맥 + 질문 포함
        input_text = f"문맥: {context} 질문: {row['input_text']} 대책을 제안해줘."
        output_text = row['answer']

        # 토크나이징
        tokenized = self.tokenizer(
            input_text, padding='max_length', truncation=True, max_length=self.max_length, return_tensors='pt'
        )
        labels = self.tokenizer(
            output_text, padding='max_length', truncation=True, max_length=self.max_length, return_tensors='pt'
        )

        return {
            'input_ids': tokenized['input_ids'].squeeze(),
            'attention_mask': tokenized['attention_mask'].squeeze(),
            'labels': labels['input_ids'].squeeze()
        }

# ==============================
# 3. KoBART 학습 모듈
# ==============================
class KoreanKoBARTTrainer:
    def __init__(self, model_name='hyunwoongko/kobart', max_length=512):
        self.tokenizer = PreTrainedTokenizerFast.from_pretrained(model_name)
        self.model = BartForConditionalGeneration.from_pretrained(model_name).to("cuda")  # GPU로 모델 이동
        self.max_length = max_length

    def fine_tune(self, train_data, output_dir='./kobart_finetuned(10000train_with_pdf)', epochs=3, batch_size=8):
        dataset = KoBARTDataset(train_data, self.tokenizer, max_length=self.max_length)

        training_args = TrainingArguments(
            output_dir=output_dir,
            num_train_epochs=epochs,
            per_device_train_batch_size=batch_size,
            save_steps=500,
            save_total_limit=2,
            logging_dir=f'{output_dir}/logs',
            logging_steps=100,
            fp16=True,
        )

        self.model.to("cuda")  # GPU로 모델 이동
        trainer = Trainer(
            model=self.model,
            args=training_args,
            train_dataset=dataset,
            data_collator=DataCollatorForSeq2Seq(self.tokenizer, model=self.model)
        )
        trainer.train()
        self.model.save_pretrained(output_dir)
        self.tokenizer.save_pretrained(output_dir)
        print(f'Model saved to {output_dir}')

    def load_finetuned_model(self, model_path):
        self.tokenizer = PreTrainedTokenizerFast.from_pretrained(model_path)
        self.model = BartForConditionalGeneration.from_pretrained(model_path).to("cuda")  # GPU 이동
        print(f'Model loaded from {model_path}')

    def generate(self, query, context, num_beams=5, max_length=None):
        # query와 context를 결합하여 모델에 입력
        input_text = f"문맥: {context} 질문: {query} 대책을 제안해줘."

        max_length = max_length or self.max_length
        inputs = self.tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True).to("cuda")  # GPU 이동
        generated_ids = self.model.generate(
            input_ids=inputs["input_ids"],
            num_beams=num_beams,
            max_length=max_length,
            early_stopping=True
        )
        output = self.tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
        return output

# ==============================
# 4. 최종 실행 코드
# ==============================
import random

# 데이터 로드 및 전처리
train_path = "train_cleaned.csv"
processor = DataProcessor()
train_data = processor.load_train_data(train_path)

# 질문과 대답을 바탕으로 fine-tuning 하기 위한 텍스트 준비
train_data["input_text"] = train_data['question']  # 'query'를 'input_text'로 사용
train_data = train_data[['input_text', 'PDF 내 관련 문장', 'answer']]  # 모델이 사용할 두 개의 컬럼만 남김

# Trainer 초기화 및 파인튜닝
trainer = KoreanKoBARTTrainer()
random_numbers = random.sample(range(1, len(train_data)-1), 10000)  # 랜덤정수 10000개 선정
trainer.fine_tune(train_data.loc[random_numbers], epochs=3, batch_size=8)  # 전체 모델 파인튜닝

# 모델 테스트
test_query = train_data.loc[0, 'input_text']  # 테스트할 질문
context = train_data.loc[0, 'PDF 내 관련 문장']  # 해당 질문에 대한 PDF 내 문맥
response = trainer.generate(test_query, context)  # 문맥을 포함한 답변 생성
print("Generated Response:", response)
```

### 데이터 후처리

```python
def clean_words(text):
  if '작업 자' in text:
    text = text.replace("작업 자","작업자")
  if '실시 를' in text:
    text = text.replace("실시 를","실시를")
  return text

for idx, i in enumerate(test_responses):
  test_responses[idx] = clean_words(i)
  print(test_responses[idx])
```

```python
from sentence_transformers import SentenceTransformer

# Embedding Vector 추출에 활용할 모델(jhgan/ko-sbert-sts) 불러오기
model = SentenceTransformer('jhgan/ko-sbert-sts', use_auth_token=False)

res_v = {}
for k,v in pd.DataFrame(test_responses).items():
    res_v[k] = model.encode(v)
```

```python
for i in range(len(test_responses)):
  sample.loc[i,'vec_0':] = res_v[0][i]
```
